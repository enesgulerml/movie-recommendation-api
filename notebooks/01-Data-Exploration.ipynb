```python
# --- 1. ratings.dat ---
r_cols = ['UserID', 'MovieID', 'Rating', 'Timestamp']
ratings_df = pd.read_csv(
    "../data/raw/ratings.dat",
    sep='::',
    header=None,
    names=r_cols,
    engine='python'
)

print("--- 1. ratings.dat (First 5 Rows) ---")
print(ratings_df.head())
print(f"\nTotal Rating: {len(ratings_df)}")
```

```python
# --- 2. movies.dat ---
m_cols = ['MovieID', 'Title', 'Genres']
movies_df = pd.read_csv(
    "../data/raw/movies.dat",
    sep='::',
    header=None,
    names=m_cols,
    engine='python',
    encoding='latin-1'
)

print("\n\n--- 2. movies.dat (First 5 Lines) ---")
print(movies_df.head())
print(f"\nTotal Movie: {len(movies_df)}")
```

```python
# --- 3. users.dat ---
u_cols = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code']
users_df = pd.read_csv(
    "../data/raw/users.dat",
    sep='::',
    header=None,
    names=u_cols,
    engine='python'
)

print("\n\n--- 3. users.dat (First 5 Lines) ---")
print(users_df.head())
print(f"\nTotal Users: {len(users_df)}")
```
